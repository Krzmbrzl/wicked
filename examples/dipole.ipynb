# Load pywicked, setup orbital spacesm, and create WickTheorem object

```python
import sys
sys.path.insert(0, '/Users/fevange/Source/wicked')
import pywicked as pw
from IPython.display import display, Math, Latex

pw.osi.reset()
pw.osi.add_space("v",pw.rdm.unoccupied,["e","f"])
pw.osi.add_space("c",pw.rdm.occupied,["m","n"])
pw.osi.add_space("a",pw.rdm.general,["u","v","w","x","y","z"])
wdt = pw.WDiagTheorem()
```

# Create operators

```python
spaces = {"c": ["c"],
          "a": ["a"],
          "v": ["v"],
          "h": ["c", "a"],
          "p": ["a", "v"],
          "g": ["c", "a", "v"]
          }

# generate list of indices
def gen_ex(ss):
    out = []
    for cspace in ss:
        temp = []
        for space in spaces[cspace]:
            if out:
                for partial_space in out:
                    temp.append(partial_space + space)
            else:
                temp.append(space)
        out = temp
    return out

# list of indices
iT1 = gen_ex("hp")
iT2 = gen_ex("hhpp")
iM  = gen_ex("gg")
iT1c = gen_ex("ph")
iT2c = gen_ex("pphh")

# remove active only excitations
iT1.remove("aa")
iT2.remove("aaaa")
iT1c.remove("aa")
iT2c.remove("aaaa")

# format list
def format_list(l):
    n = len(l[0])
    half = n / 2
    out = []
    for i in l:
        out.append(i[:half] + "->" + i[half:])
    return out

iT1 = format_list(iT1)
iT2 = format_list(iT2)
iM  = format_list(iM)
iT1c = format_list(iT1c)
iT2c = format_list(iT2c)

# combine T1 and T2
iA = [i for i in iT1]
iA.extend(iT2)
# iA.extend(iT1_)
# iA.extend(iT2_)

T1 = pw.make_operator("T", iT1)
T2 = pw.make_operator("T", iT2)
T1c = pw.make_operator("T", iT1c)
T2c = pw.make_operator("T", iT2c)
A = pw.make_operator("A", iA)
M = pw.make_operator("M", iM)

M = []
for cre in ["c", "a", "v"]:
    for ann in ["c", "a", "v"]:
        M.append(pw.make_diag_operator("M", [cre], [ann]))
      
# Mcc = pw.make_diag_operator("M", ["c"], ["c"])
# Mac = pw.make_diag_operator("M", ["c"], ["a"])
# Mca = pw.make_diag_operator("M", ["a"], ["c"])
# Mvc = pw.make_diag_operator("M", ["c"], ["v"])
# Mcv = pw.make_diag_operator("M", ["v"], ["c"])
# Maa = pw.make_diag_operator("M", ["a"], ["a"])
# Mva = pw.make_diag_operator("M", ["a"], ["v"])
# Mav = pw.make_diag_operator("M", ["v"], ["a"])
# Mvv = pw.make_diag_operator("M", ["v"], ["v"])

T1 = []
T2 = []
T1Hc = []
T2Hc = []
A = []
B = []
for cre in ["a", "v"]:
    for ann in ["c", "a"]:
        if cre == "a" and ann == "a":
            continue
        T1.append(pw.make_diag_operator("T", [cre], [ann]))
        T1Hc.append(pw.make_diag_operator("T", [ann], [cre]))
        A.append(pw.make_diag_operator("A", [cre], [ann]))
        A.append(pw.make_diag_operator("A", [ann], [cre]))
        B.append(pw.make_diag_operator("B", [cre], [ann]))
        B.append(pw.make_diag_operator("B", [ann], [cre]))
        
for cre0 in ["a", "v"]:
    for cre1 in ["a", "v"]:
        for ann0 in ["c", "a"]:
            for ann1 in ["c", "a"]:
                if cre0 == "a" and cre1 == "a" and ann0 == "a" and ann1 == "a":
                    continue
                T2.append(pw.make_diag_operator("T", [cre0, cre1], [ann0, ann1]))
                T2Hc.append(pw.make_diag_operator("T", [ann0, ann1], [cre0, cre1]))
                A.append(pw.make_diag_operator("A", [cre0, cre1], [ann0, ann1]))
                A.append(pw.make_diag_operator("A", [ann0, ann1], [cre0, cre1]))
                B.append(pw.make_diag_operator("B", [cre0, cre1], [ann0, ann1]))
                B.append(pw.make_diag_operator("B", [ann0, ann1], [cre0, cre1]))

print T1
print T1Hc
# F = pw.make_operator("F", ["o->o","o->v","v->o","v->v"]);
# V = pw.make_operator("V", ["oo->oo", "oo->ov", "oo->vv",
#                            "ov->oo", "ov->ov", "ov->vv",
#                            "vv->oo", "vv->ov", "vv->vv"]);
# H = pw.make_operator("v", ["o->o"  , "o->v"  , "v->o"  , "v->v",
#                            "oo->oo", "oo->ov", "oo->vv",
#                            "ov->oo", "ov->ov", "ov->vv",
#                            "vv->oo", "vv->ov", "vv->vv"]);
```
